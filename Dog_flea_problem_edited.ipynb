```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
import random
```

```python
N=1000
dogA=np.arange(1,N+1,1)
#print(dogA)
dogB=[]
N_A=[]
N_B=[]
step=0.1
t_final=5000
t_list=np.arange(0.1,t_final,step)

def entropy(N,m):
    n_C_m=math.factorial(N)/(math.factorial(m)*math.factorial(N-m))
    S= math.log(n_C_m)
    return S

S_list=[]
for t in t_list:
    q= random.randrange(1, N+1, 1)
    for i in range (0,len(dogA)):
        if(dogA[i]==q):
            dogB_n=np.append(dogB,dogA[i])
            dogA_n=np.delete(dogA,i)
    for z in range (0,len(dogB)):
        if(dogB[z]==q):
            dogB_n=np.delete(dogB,z)
            dogA_n=np.append(dogA,dogB[z])
    dogB=np.array(dogB_n)
    dogA=np.array(dogA_n)
    #print(dogA)
    m=len(dogA)
    N_A.append(len(dogA))
    N_B.append(len(dogB))
    S_list.append(entropy(N,m))


N_A=np.array(N_A)
N_B=np.array(N_B)
sum_list=N_A+N_B
    #print(N_B)
    
```

```python
plt.scatter(t_list,N_A,label="flea on dog A")
plt.scatter(t_list,N_B,label="flea on dog B")
plt.scatter(t_list,sum_list,label="total flea")
plt.legend()
plt.xlim(-100,2000)
```

```python
plt.plot(t_list,S_list,label="entropy")
plt.legend()
```
